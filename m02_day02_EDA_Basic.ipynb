```python
import pandas as pd

data = {
    "height": [165, 172, None, 180, 158, 175],
    "weight": [55, 70, 65, 80, 50, 68],
    "age": [25, 30, 28, 35, 22, 29],
    "gender": ["F", "M", "M", "M", "F", "M"],
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum() # 결측치 개수 파악 가능
```

```python
# 결측치가 있는 행 제거
df2 = df.dropna()
df2
```

```python
df3 = df.copy() # df복사 --> df3로
```

```python
# 결측치 채우기 --> fillna()
# 숫자형 --> 평균값ㅇ로 채우기
df3['height'] = df['height'].fillna(df['height'].mean()) # 덮어씌울려면 이렇게 해야 된다~
df3
```

```python
# 결측치 채우기 --> fillna()
# 범주형 결측치를 최빈값으로 채우기 # 문자형일 때는 최빈값을 하는게 reasonable
df['gender'].mode()
```

# 인코딩 (Encoding)
- 문자 데이터를 숫자로 바꾸는 과정

## Label Encoding
- 값에 숫자를 직접 매긴다.

```python
df
```

```python
df['gender_num'] = df['gender'].map({'F':0, 'M':1}) # map()안에는 딕셔너리 형태로
df
```

## One-Hot Encoding
- True / False
- 1 / 0
- 참거짓, 성별 같이 이분법적인 개념들에 주로 쓰이는 방식

```python
df_encoded = pd.get_dummies(df, columns=['gender'])
df_encoded
```

# 스케일링
- 숫자 값의 범위를 조정
- ex. 나이 --> 20~70, 연봉 --> 30,000,000

## MinMaxScaler
- 사이킷런 안에 있음
- 값을 0 ~ 1 범위로 바꾸어준다

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['height_scaled']] = scaler.fit_transform(df[['height']])
```

```python
df   # 제일 큰 애가 1, 작은 애가 0
```

## StandardScaler
- 평균이 0, 표준편차가 1이 되도록 조정
- scikit learn 안에 있음

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['height_standard']] = scaler.fit_transform(df[['height']])
df
```
